```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/credit_card_approval_dataset.csv")
```

```python
counts = df['Credit Status'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(counts.index, counts.values, color=['red', 'green'])
plt.title("Loan Approval Count")
plt.xlabel("Credit Status")
plt.ylabel("Number of Applicants")
plt.show()
```

```python
approved = df[df['Credit Status'] == 'Approved']['Salary']
rejected = df[df['Credit Status'] == 'Rejected']['Salary']

plt.figure(figsize=(6,4))
plt.boxplot([approved, rejected], tick_labels=['Approved', 'Rejected'])

plt.title("Salary vs Credit Status")
plt.xlabel("Credit Status")
plt.ylabel("Income")

plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['Salary'], bins=20, color='blue', edgecolor='black')

plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Salary'], df['Income'], alpha=0.5, color='pink')

plt.title("Salary vs Income")
plt.xlabel("Salary")
plt.ylabel("Income")

plt.show()
```
